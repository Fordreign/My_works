5 Сегментация

```python
df.columns
```

```python
df5 = df.loc[(df['event_id_1'].isin(['horde', 'ctf', 'koth', 'dm', 'tdm']))][['session_id','event_id_1','diff_date']].groupby(['session_id'])[['event_id_1','diff_date']].min()
```

```python
df5.head()
```

```python
df5['event_id_1'] = df5['event_id_1'].astype('str')
```

```python
df5.info()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Выбираем интересующие нас колонки
df5_subset = df5[['event_id_1', 'diff_date']]

# Преобразуем колонку diff_date в число секунд
df5_subset['diff_sec'] = df5_subset['diff_date'].dt.total_seconds()

# Масштабируем данные
scaler = StandardScaler()
df5_scaled = scaler.fit_transform(df5_subset[['diff_sec']])

# Применяем алгоритм кластеризации KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df5_subset['cluster'] = kmeans.fit_predict(df5_scaled)

# Выводим результаты кластеризации
print(df5_subset.groupby(['event_id_1', 'cluster'])['diff_sec'].describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Выбираем интересующие нас колонки
df5_subset = df5[['event_id_1', 'diff_date']]

# Преобразуем колонку diff_date в число секунд
df5_subset['diff_sec'] = df5_subset['diff_date'].dt.total_seconds()

# Масштабируем данные
scaler = StandardScaler()
df5_scaled = scaler.fit_transform(df5_subset[['diff_sec']])

# Применяем алгоритм кластеризации KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df5_subset['cluster'] = kmeans.fit_predict(df5_scaled)

# Определяем цвет и форму точек для каждого кластера
palette = sns.color_palette("bright", len(df5_subset['cluster'].unique()))
markers = ['o', 's', '^', 'd']

# Строим график
sns.scatterplot(data=df5_subset, x='event_id_1', y='diff_sec', hue='cluster', palette=palette, style='cluster', markers=markers)
plt.xlabel('Игровой режим')
plt.ylabel('Время игры (секунды)')
plt.title('Сегментация по времени игры и игровым режимам')
plt.show()
```

```python
df5_subset.columns
```

```python
# Строим boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df5_subset, x='event_id_1', y='diff_sec', hue='cluster')
plt.xlabel('Игровой режим')
plt.ylabel('Время игры (секунды)')
plt.title('Сегментация по времени игры и игровым режимам')
plt.show()
```
