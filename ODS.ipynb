```python
with open('/Users/PC/Desktop/DATA_SCIENCE/DATA/input_file.txt') as f:
    my_lines = f.readlines()  
def cd(list):
    d=[]
    for i in list:
        a = i.split()
        c = eval('{1}{0}{2}'.format(*a))
        d.append(c)
    return d
ans_file = open('output_ttt.csv', 'w')
for _ in cd(my_lines):
    ans_file.write(str(_))
    ans_file.write(',')
ans_file.close()
```
