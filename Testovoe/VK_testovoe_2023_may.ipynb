Посторим график для средних gvm по аккаунтам и HID из доступных данных для компаний Green, Blue и Black.

```python
# Создание списка средних GMV для каждой компании
gvm_means = [df['Green_gvm_total'].sum()/df.query('has_Green_id > 0').shape[0],
             df['Blue_gvm_total'].sum()/df.query('has_Blue_id > 0').shape[0],
             df['Black_gvm_total'].sum()/df.query('has_Black_id > 0').shape[0]]

# Создание списка названий компаний и их цветов
companies = ['Green', 'Blue', 'Black']
colors = ['green', 'blue', 'black']

# Построение графика
fig, ax = plt.subplots()
ax.bar(companies, gvm_means, color=colors)
ax.set_xlabel('Компания')
ax.set_ylabel('Средний GVM на пользователя HID')
ax.set_title('Средний GVM на пользователя HID')

# Добавление текста с значениями средних GVM на пользователя HID для каждой компании
for i, v in enumerate(gvm_means):
    ax.text(i, v, f"{v:.2f} у.е.", ha='center', fontsize=12)

plt.show()
```

```python
# Создание списка средних GMV для каждой компании
gvm_means = [df['Green_gvm_total'].sum()/df['has_Green_id'].sum(),
             df['Blue_gvm_total'].sum()/df['has_Blue_id'].sum(),
             df['Black_gvm_total'].sum()/df['has_Black_id'].sum()]

# Создание списка названий компаний и их цветов
companies = ['Green', 'Blue', 'Black']
colors = ['green', 'blue', 'black']

# Построение графика
fig, ax = plt.subplots()
ax.bar(companies, gvm_means, color=colors)
ax.set_xlabel('Компания')
ax.set_ylabel('Средний GVM на аккаунт')
ax.set_title('Средний GVM на аккаунт')

# Добавление текста с значениями средних GVM на аккаунт для каждой компании
for i, v in enumerate(gvm_means):
    ax.text(i, v, f"{v:.2f} у.е.", ha='center', fontsize=12)

plt.show()
```

Теперь посмотрим каким устройством чаще пользуются внутри каждой компании

```python
# Создание списка с названиями компаний и устройств
companies = ['Red', 'Green', 'Blue', 'Black']
devices = ['android', 'apple', 'web']

# Создание пустого словаря для хранения данных о частоте использования каждого устройства для каждой компании
device_counts = {}

# Заполнение словаря данными
for company in companies:
    device_counts[company] = []
    for device in devices:
        col_name = f"{company}_{device}"
        device_counts[company].append(df[col_name].sum())

# Построение графиков
fig, ax = plt.subplots(figsize=(10, 8))
for i, company in enumerate(companies):
    plt.subplot(2, 2, i+1)
    plt.pie(device_counts[company], labels=devices, autopct='%1.1f%%', colors=['orange', 'purple', 'brown'])
    plt.title(company)
    plt.axis('equal')

plt.show();
```

Как видно из графиков пользователи Red компании преподчитают web устройство остальным, а вот пользователи Green и Black компании чаще используют apple устройства, у Blue компании клиенты почти распределы одинаково, чуть чаще используют web, чем apple устройства.

И наконец узнаем, как влияет is_msk на наши данные.

```python
print(f"Среднее количество аккаунтов на пользователя не из Москвы для каждой компании :")
print(f"Компания Red: {round(df.query('is_msk == 0')['has_Red_id'].sum() / df.query('has_Red_id > 0 & is_msk == 0').shape[0], 3)}")
print(f"Компания Green: {round(df.query('is_msk == 0')['has_Green_id'].sum() / df.query('has_Green_id > 0 & is_msk == 0').shape[0], 3)}")
print(f"Компания Black: {round(df.query('is_msk == 0')['has_Black_id'].sum() / df.query('has_Black_id > 0 & is_msk == 0').shape[0], 3)}")
print(f"Компания Blue: {round(df.query('is_msk == 0')['has_Blue_id'].sum() / df.query('has_Blue_id > 0 & is_msk == 0').shape[0], 3)}")

print(f"Среднее количество аккаунтов на пользователя из Москвы для каждой компании :")
print(f"Компания Red: {round(df.query('is_msk == 1')['has_Red_id'].sum() / df.query('has_Red_id > 0 & is_msk == 1').shape[0], 3)}")
print(f"Компания Green: {round(df.query('is_msk == 1')['has_Green_id'].sum() / df.query('has_Green_id > 0 & is_msk == 1').shape[0], 3)}")
print(f"Компания Black: {round(df.query('is_msk == 1')['has_Black_id'].sum() / df.query('has_Black_id > 0 & is_msk == 1').shape[0], 3)}")
print(f"Компания Blue: {round(df.query('is_msk == 1')['has_Blue_id'].sum() / df.query('has_Blue_id > 0 & is_msk == 1').shape[0], 3)}")
```

```python
# Создание списка средних GMV для каждой компании в Москве
gvm_means_msk = [df.query('is_msk == 1')['Green_gvm_total'].sum()/df.query('has_Green_id > 0 & is_msk == 1').shape[0],
                 df.query('is_msk == 1')['Blue_gvm_total'].sum()/df.query('has_Blue_id > 0 & is_msk == 1').shape[0],
                 df.query('is_msk == 1')['Black_gvm_total'].sum()/df.query('has_Black_id > 0 & is_msk == 1').shape[0]]

# Создание списка средних GMV для каждой компании вне Москвы
gvm_means_other = [df.query('is_msk == 0')['Green_gvm_total'].sum()/df.query('has_Green_id > 0 & is_msk == 0').shape[0],
                   df.query('is_msk == 0')['Blue_gvm_total'].sum()/df.query('has_Blue_id > 0 & is_msk == 0').shape[0],
                   df.query('is_msk == 0')['Black_gvm_total'].sum()/df.query('has_Black_id > 0 & is_msk == 0').shape[0]]

# Создание списка названий компаний и их цветов
companies = ['Green', 'Blue', 'Black']
colors = ['green', 'blue', 'black']

# Построение графика
fig, ax = plt.subplots(figsize=(8,6))
bar_width = 0.35
opacity = 0.8
index = np.arange(len(companies))

# Графики для средних GMV в Москве и вне Москвы
ax.bar(index, gvm_means_msk, bar_width, alpha=opacity, color='purple', label='Москва')
ax.bar(index + bar_width, gvm_means_other, bar_width, alpha=opacity, color='lightgray', label='Другие регионы')

ax.set_xlabel('Компания')
ax.set_ylabel('Средний GVM на пользователя HID')
ax.set_title('Средний GVM на пользователя HID по признаку is_msk')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(companies)
ax.legend()

# Добавление текста с значениями средних GVM на пользователя HID для каждой компании
for i, v in enumerate(gvm_means_msk):
    ax.text(i, v, f"{v:.2f} у.е.", ha='center', fontsize=12, color='black')
for i, v in enumerate(gvm_means_other):
    ax.text(i+bar_width, v, f"{v:.2f} у.е.", ha='center', fontsize=12)

plt.show()
```

Для Black компании в среднем почти в 2 раза больше приносят GVM пользователи из Москвы, для Green компании уже не так сильна эта разница, примерно на 27% , что тоже ощутимо.

Посмотрим теперь посмотрим, где чаще пользуются клиенты каждой компании и каким устройством

```python
# Создание списка с названиями компаний и устройств
companies = ['Red', 'Green', 'Blue', 'Black']
devices = ['android', 'apple', 'web']

# Создание пустого словаря для хранения данных о частоте использования каждого устройства для каждой компании
device_counts_msk = {}
device_counts_not_msk = {}

# Заполнение словарей данными
for company in companies:
    device_counts_msk[company] = []
    device_counts_not_msk[company] = []
    for device in devices:
        col_name = f"{company}_{device}"
        device_counts_msk[company].append(df.query('is_msk==1')[col_name].sum())
        device_counts_not_msk[company].append(df.query('is_msk==0')[col_name].sum())

# Построение графиков для Москвы
fig, ax = plt.subplots(figsize=(10, 8))
for i, company in enumerate(companies):
    plt.subplot(2, 2, i+1)
    plt.pie(device_counts_msk[company], labels=devices, autopct='%1.1f%%', colors=['orange', 'purple', 'brown'])
    plt.title(f"{company} in Moscow")
    plt.axis('equal')

plt.show();

# Построение графиков для других регионов
fig, ax = plt.subplots(figsize=(10, 8))
for i, company in enumerate(companies):
    plt.subplot(2, 2, i+1)
    plt.pie(device_counts_not_msk[company], labels=devices, autopct='%1.1f%%', colors=['orange', 'purple', 'brown'])
    plt.title(f"{company} in other regions")
    plt.axis('equal')

plt.show();
```

В регионах клиенты чаще пользуются web , а вот в компании Black наоборот, чаще в Москве.
