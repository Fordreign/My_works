Data structuring

```python
import json
```

```python
df['event_json'] = df['event_json'].apply(lambda x: json.loads(x))
```

```python
df['event_json'][0]
```

```python
df['ad_type'] = df['event_json'].apply(lambda x: x['ad_type'])
```

```python
df['revenue'] = df['event_json'].apply(lambda x: x['revenue'])
```

```python
df['network_name'] = df['event_json'].apply(lambda x: x['network_name'])
```

```python
df.columns
```

```python
df_new = df[['profile_id', 'os_version','city', 'ad_type', 'revenue', 'network_name']]
df_new.head()
```

```python
df_new.nunique()
```

```python
df_new['city'].fillna('Unknow', inplace=True)
```

eCPM calculation + graphics

```python
'''
Посмотрим сколько в общем итоге мы заработали на каждом типе рекламы.
'''
df_new.groupby('ad_type')['revenue'].sum()
```

```python
'''
Взгляним на количество рекламы каждого типа.
'''
df_new.groupby('ad_type').size()
```

```python
'''
Расчет eCPM.
'''
df_new.groupby('ad_type')['revenue'].mean() * 1000
```

```python
banner_mean = df_new.loc[df_new['ad_type'] == 'banner'].groupby('profile_id')['revenue'].mean() #распр. средних eCPM(banner) юзеров
int_mean = df_new.loc[df_new['ad_type'] == 'int'].groupby('profile_id')['revenue'].mean() #распр. средних eCPM(int) юзеров
rv_mean = df_new.loc[df_new['ad_type'] == 'rv'].groupby('profile_id')['revenue'].mean() #распр. средних eCPM(rc) юзеров
```

```python
sns.kdeplot(rv_mean).set(title='rv_mean');
print(f'Среднее : {np.round(rv_mean.mean(),4)}\nСр.кв.отклонение : {np.round(rv_mean.var(),4)}\neCPM : 65.051165')
```

```python
sns.kdeplot(int_mean).set(title='int_mean');
print(f'Среднее {np.round(int_mean.mean(),4)}\nСр.кв.отклонение {np.round(int_mean.var(),4)}\neCPM : 36.209634')
```

```python
sns.kdeplot(banner_mean).set(title='banner_mean');
print(f'Среднее : {np.round(banner_mean.mean(),4)}\nСр.кв.отклонение : {np.sqrt(np.round(banner_mean.var(),6))}\neCPM : 0.744113')
```

'''
Взглянув на графики , можно заметить у всех распр. тяжелые хвосты, возможо это зависит от рекламной сети или региона.
По среднему ил ср.кв.откл. можно предварительно заключить , что от типа рекламы зависит эти параметры.
'''

```python
plt.figure(figsize=(8, 4), dpi=80)
df_new.groupby('os_version')['revenue'].mean().sort_values(ascending=False).plot(kind='bar')
plt.ylabel('Общий доход')
plt.title('Средний eCPM  по версиям ОС')
plt.show;
```

```python
plt.figure(figsize=(8, 4), dpi=80)
df_new.groupby('os_version')['revenue'].sum().sort_values(ascending=False).plot(kind='bar')
plt.ylabel('Общий доход')
plt.title('Суммарный доход по версия ОС')
plt.show;
```

'''
Основной приток дохода на более новые версии ОС.
Эта информация будет полезна для разработчиков.
'''

```python
plt.figure(figsize=(8, 6), dpi=80)
df_new.groupby('city')['revenue'].sum().sort_values(ascending=False).head(30).plot(kind='barh')
plt.xlabel('Общий доход')
plt.ylabel('Город')
plt.title('Суммарный доход по городам')
plt.show;
```

```python
"""
Давайте посчитаем какой процент дохода содержит информацию о городе юзера.
"""
df_new.loc[df_new['city'] == 'Unknow']['revenue'].sum() / df_new.groupby('city')['revenue'].sum().sum()
```

'''
Треть наших юзеров неизвестны нам по городу.
Посмоторим на то , что есть.
'''

```python
plt.figure(figsize=(8, 6), dpi=80)
df_new.groupby('city')['revenue'].sum().sort_values(ascending=False).iloc[1:30].plot(kind='barh')
plt.xlabel('Общий доход')
plt.ylabel('Город')
plt.title('Суммарный доход по городам')
plt.show;
```

```python
plt.figure(figsize=(8, 6), dpi=80)
df_new.groupby('city')['revenue'].mean().sort_values(ascending=False).iloc[1:30].plot(kind='barh')
plt.xlabel('Общий доход')
plt.ylabel('Город')
plt.title('Средний eCPM  по городам')
plt.show;
```

'''
По графика видно , что самые прибыльные города(в частоности и более населенные) не имеют высокого среднего eCPM.
'''

```python
plt.figure(figsize=(8, 6), dpi=80)
df_new.groupby('network_name')['revenue'].sum().sort_values(ascending=False).plot(kind='barh')
plt.xlabel('Общий доход')
plt.ylabel('Рекламная сеть')
plt.title('Суммарных доход от рекламной сети')
plt.show;
```
